```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
data = pd.DataFrame({
    'Values': np.random.normal(loc=0, scale=1, size=1000),
    'Category': np.random.choice(['A', 'B', 'C'], size=1000),
    'X': np.random.normal(loc=0, scale=1, size=1000),
    'Y': np.random.normal(loc=0, scale=1, size=1000),
    'Date': pd.date_range(start='2023-01-01', periods=1000),
})
```

```python
# Distributions
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(data['Values'], kde=True)
plt.title('Histogram with KDE')
```

```python
# Categorical Distributions
plt.subplot(1, 2, 2)
sns.countplot(data['Category'])
plt.title('Countplot')
plt.tight_layout()
plt.show()
```

```python
# 2D Distributions (using hexbin plot from Matplotlib)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(data=data, x='X', y='Y', alpha=0.5)
plt.title('Scatter Plot')

plt.subplot(1, 2, 2)
plt.hexbin(data['X'], data['Y'], gridsize=20, cmap='coolwarm')
plt.colorbar(label='count')
plt.title('Hexbin Plot')

plt.tight_layout()
plt.show()
```

```python
# Time Series
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.lineplot(data=data, x='Date', y='Values')
plt.title('Time Series Line Plot')

plt.subplot(1, 2, 2)
sns.lineplot(data=data, x='Date', y='Values', estimator='mean')
plt.title('Time Series with Mean Line')

plt.tight_layout()
plt.show()
```

```python
# Faceted Distributions
g = sns.FacetGrid(data, col='Category', height=4)
g.map(sns.histplot, 'Values', bins=30)
plt.show()
```

Importing Libraries



```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train.shape
```

```python

y_train.shape
```

```python
k=9
knclr=KNeighborsClassifier(k)
```

```python
#Training the Model


knclr
```

```python
knclr.fit(x_train,y_train)
```

```python
y_pred=knclr.predict(x_test)


metrics.accuracy_score(y_test, y_pred)*100
```
